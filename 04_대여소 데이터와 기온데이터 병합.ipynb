```python
import pandas as pd
import datetime
```

```python
# 2023년 대여소별 자전거 대여건수 데이터와 2023년 날씨+미세먼지 데이터 불러오기
station_rental_2023 = pd.read_csv('2023년 대여소별 공공자전거 대여건수(시간대별).csv')
weather_dust_2023 = pd.read_csv('[결측치 제거]2023년 날씨+미세먼지정보.csv', encoding='cp949')
```

```python
station_rental_2023
```

```python
# 데이터프레임 병합을 위해 열 이름 통일
weather_dust_2023.rename(columns= {'날짜' : '대여일자'}, inplace=True)
weather_dust_2023.rename(columns= {'시간대' : '대여시간'}, inplace=True)
```

```python
weather_dust_2023[:48]
```

```python
# 2023년 대여소별 대여건수 데이터를 기준으로 두 데이터 병합
station_weather_rental_2023 = pd.merge(station_rental_2023, weather_dust_2023, on=['대여일자', '대여시간'], how = 'left')
station_weather_rental_2023[:24]
```

```python
# 결측치 확인
station_weather_rental_2023.isnull().sum()
```

```python
station_weather_rental_2023.to_csv('2023년 대여소별 대여 및 기온데이터.csv', index=False, encoding='cp949')
```
