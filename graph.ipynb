```python
import pandas as pd
import matplotlib.pyplot as plt

plt.rc('font', family='batang') # For Windows
print(plt.rcParams['font.family'])

data = pd.read_csv(r'C:\Users\vuswl\Data Science\2023년 따릉이 대여 및 날씨 데이터.csv', engine = 'python', encoding='CP949')
data.head()
```

```python
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np

# 01. 분포
mpl.rc('font', size = 15)
sns.displot(np.log(data['이용건수']))
```

```python
print(data.head())
```

```python
# 시간대별 이용건수 막대그래프
mpl.rc('font', size=7)
mpl.rc('axes', titlesize=15)
sns.barplot(x='시간대', y='이용건수', data=data)
```

```python
# 시간대별 이용건수 boxplot
sns.boxplot(x='대여월', y='이용건수', data=data)
```

```python
# 평일 여부에 따른 시간대별 이용건수 (포인트플롯)
sns.pointplot(x = "시간대", y = "이용건수", data = data, hue = "평일", markersize = 3)
```

```python
# 시간대에 따른 대여월별 이용건수 (포인트플롯)
sns.pointplot(x = "대여월", y = "이용건수", data = data, hue="시간대", markersize = "3")
```

```python
# 회귀선을 포함한 산점도 그래프

# 2행 3열으로 Figure 설정
mpl.rc('font', size = 15)
figure, axes = plt.subplots(nrows = 2, ncols = 3)
plt.tight_layout()
figure.set_size_inches(10, 5)

# 서브플롯 할당
# 기온, 강수량, 풍속, 습도, 적설, 미세먼지농도, 대여월, 평일, 시간대
sns.regplot(x = "기온(°C)", y = "이용건수", data = data, ax = axes[0, 0], scatter_kws = {'alpha' : 0.2}, line_kws = {'color' : 'orange'})
sns.regplot(x = "강수량(mm)", y = "이용건수", data = data, ax = axes[0, 1], scatter_kws = {'alpha' : 0.2}, line_kws = {'color' : 'orange'})
sns.regplot(x = "풍속(m/s)", y = "이용건수", data = data, ax = axes[0, 2], scatter_kws = {'alpha' : 0.2}, line_kws = {'color' : 'orange'})
sns.regplot(x = "습도(%)", y = "이용건수", data = data, ax = axes[1, 0], scatter_kws = {'alpha' : 0.2}, line_kws = {'color' : 'orange'})
sns.regplot(x = "적설(cm)", y = "이용건수", data = data, ax = axes[1, 1], scatter_kws = {'alpha' : 0.2}, line_kws = {'color' : 'orange'})
sns.regplot(x = "미세먼지농도(㎍/㎥)", y = "이용건수", data = data, ax = axes[1, 2], scatter_kws = {'alpha' : 0.2}, line_kws = {'color' : 'orange'})
```

```python
data[['기온(°C)', '강수량(mm)', '풍속(m/s)', '습도(%)', '적설(cm)', '미세먼지농도(㎍/㎥)', '평일', '시간대', '이용건수']].corr()
```

```python
corrMat = data[['기온(°C)', '강수량(mm)', '풍속(m/s)', '습도(%)', '적설(cm)', '미세먼지농도(㎍/㎥)',  '평일', '시간대', '이용건수']].corr()
fig, ax=plt.subplots()
fig.set_size_inches(10, 10)
sns.heatmap(corrMat, annot=True)
ax.set(title='Heatmap of Numerical Data');
```

```python
import matplotlib.pyplot as plt

plt.title('대여월별 이용건수')
plt.xlabel('대여월')

ax = sns.countplot(x = "대여월", data = data, color = "skyblue")

ax2 = ax.twinx()
ax2 = sns.pointplot(x = "대여월", y = "이용건수", color = 'pink', data = data)

plt.ylabel('이용건수', color='black')
plt.show()

# -----
plt.title('평일여부에 따른 이용건수')
plt.xlabel('평일 여부')

ax = sns.countplot(x = "평일", data = data, color = "skyblue")

ax2 = ax.twinx()
ax2 = sns.pointplot(x = "평일", y = "이용건수", color = 'pink', data = data)

plt.ylabel('이용건수', color='black')
plt.show()
# -----
plt.title('시간대별 이용건수')
plt.xlabel('시간대')

ax = sns.countplot(x = "시간대", data = data, color = "skyblue")

ax2 = ax.twinx()
ax2 = sns.pointplot(x = "시간대", y = "이용건수", color = 'pink', data = data)

plt.ylabel('이용건수', color='black')
plt.show()
```
